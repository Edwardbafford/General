### Bags
Dask bags are useful for parallizing operations on python data structures such as lists. Convert a sequence into a Dask bag and many operations are now accessable with parallel execution.

```python
import dask.bag as db
import numpy as np
```

```python
bag = ['a',{},{'b':3,'c':5},'']
d_bag = db.from_sequence(bag)
d_bag.all().visualize()
```

```python
d_bag.all().compute()
```

#### FIles
Bags support text files as well :)

```python
file_bag = db.read_text('bag-test.txt')
file_bag.count().compute()
```

```python
file_bag.take(1)
```

```python
file_bag.str.upper().compute()
```
